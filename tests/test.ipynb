### As imagens example_receipt_9.jpg e example_receipt_11.jpg apresentaram um desempenho superior, no entanto, elas possuem uma quantidade mais elevada de elementos. Mesmo assim, proporcionalmente, continuam sendo as mais adequadas para as funções de REGEX que foram desenvolvidas.

```python
# Considerando apenas as colunas item e ean
df_temp = df_final_ordenado[['item', 'ean']]

# Conta quantos EANs distintos existem por item
itens_com_multiplos_eans = (
    df_temp.groupby('item')['ean']
    .nunique()
    .reset_index(name='qtd_eans')
)

# Filtra os que têm mais de 1 EAN
itens_duplicados = itens_com_multiplos_eans[itens_com_multiplos_eans['qtd_eans'] > 1]

# Recupera as linhas originais desses itens e reseta o índice
df_resultado = df_temp[df_temp['item'].isin(itens_duplicados['item'])].reset_index(drop=True)
display(df_resultado)
```

```python
df_qtd = (
    df_resultado
    .groupby('item')['ean']
    .nunique()
    .reset_index(name='qtd_eans')
    .sort_values(by='qtd_eans', ascending=False)
)

print(df_qtd)
```

```python
# Itens com 2 ou mais 2 EANs
itens_muitos_eans = df_qtd[df_qtd['qtd_eans'] >= 2]
print(itens_muitos_eans)
```
